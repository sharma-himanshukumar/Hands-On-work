```
import numpy as np
from PIL import Image
import os
import re
import nltk
from nltk.corpus import stopwords
import matplotlib.pyplot as plt

from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
nltk.download("stopwords")
nltk.download('punkt')
nltk.download('wordnet')
from nltk.stem import 	WordNetLemmatizer
```

```
data=open(r'/content/money_hiest_script.txt',mode='r',encoding='cp1252')
```

```
a=data.read()
```

```
wordnet_lemmatizer = WordNetLemmatizer()
def clean(text):
  text=text.replace("\n"," ")
  text=text.replace("\n"," ")
  text = nltk.word_tokenize(text)
  text= [re.sub('[()?.><$^''\&%!]+', '', i) for i in text]
  text=[re.sub('[\\:@#_].*|[\d]+',"",i) for i in text]
  text=[i.strip() for i in text]
  text=[i for i in text if i not in stopwords.words('english') and len(i)>3 and i.startswith(("#","@","http","//"))==False]#"#" not in i and "@" not in i and i.startswith("http://")==False ]
  text=[wordnet_lemmatizer.lemmatize(i) for i in text ]
  text=[i.strip() for i in text]
  return text
```

```
data=" ".join(clean(a))
```

```
data
```

```
char_mask = np.array(Image.open("/content/Picture2.png"))    
```

```
image_colors = ImageColorGenerator(char_mask)
wc=WordCloud(background_color="white",stopwords=STOPWORDS,max_words=1000,mask=char_mask)# width=600, height=600,random_state=1)
wc.generate(data)
plt.figure(figsize=(15,25))
plt.imshow(wc.recolor(color_func=image_colors))
plt.axis("off")
# plt.savefig('N.png')
plt.show()
```
